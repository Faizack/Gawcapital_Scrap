```python
from bs4 import BeautifulSoup
import pandas as pd
import requests

def scrape_links(url):
    response = requests.get(url)
    if response.status_code == 200:
        html_content = response.content
        soup = BeautifulSoup(html_content, 'html.parser')
        links = soup.find_all('a', class_='arrow-link')
        link_data = [(link.text.strip(), link['href']) for link in links]
        return link_data
    else:
        print(f"Failed to fetch HTML content from {url}. Status code: {response.status_code}")
        return []

def add_links_to_dataframe(df, url_column_name):
    df['Scraped_Link'] = df[url_column_name].apply(scrape_links)
    return df


# Add links to DataFrame
df = add_links_to_dataframe(df, 'Link_Title')

print(df)
```

```python
import pandas as pd
df=pd.read_excel('ReportWithCompany.xlsx')
```

```python
import ast

# Function to convert string representation of tuple to actual tuple
def convert_to_tuple(string):
    try:
        return ast.literal_eval(string)
    except ValueError:
        return None

# Function to extract links from each row
def extract_links(row):
    links_dict = {}
    for link_tuple in row:
        links_dict[link_tuple[0]] = link_tuple[1]
    return links_dict

# Convert string representation of tuples to actual tuples
df['Scraped_Link'] = df['Scraped_Link'].apply(convert_to_tuple)

# Apply the function to each row in the 'Scraped_Link' column
df['All_Links'] = df['Scraped_Link'].apply(lambda x: extract_links(x))
```

```python
df['All_Links'][10]
```

```python
# Apply the function to each row in the 'Scraped_Link' column
df_links = df['Scraped_Link'].apply(lambda x: extract_links(x))

# Create a DataFrame from the extracted links
df_links = pd.DataFrame(df_links.tolist())

# Combine the extracted links with the original DataFrame
df = pd.concat([df, df_links], axis=1)

```

```python
df
```

```python
# # Drop the original 'Scraped_Link' column
df.drop(columns=['Scraped_Link'], inplace=True)
```

```python
df.columns
```

```python
# Define the base URL
base_url = 'https://planet-tracker.org'

# Function to attach base URL if not already present
def attach_base_url(link):
    if link.startswith('/'):
        return base_url + link
    else:
        return link

# Update links in 'View Report' column
df['View Report'] = df['View Report'].apply(attach_base_url)

```
