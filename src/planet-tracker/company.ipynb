```python
import os
import pandas as pd

# Define the directory paths
source_directory = r'c:\Users\faiza\Music\greenfi\gawcapital\src\planet-tracker\DB'
destination_directory = r'c:\Users\faiza\Music\greenfi\gawcapital\src\planet-tracker\data2'

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# List all files in the source directory
files = os.listdir(source_directory)

# Loop through each file
for file in files:
    # Construct the full file paths
    source_file_path = os.path.join(source_directory, file)
    destination_file_path = os.path.join(destination_directory, file)
    
    # Check if the file is an Excel file
    if file.endswith('.xlsx') or file.endswith('.xls'):
        # Read the Excel file into a DataFrame
        df = pd.read_excel(source_file_path)
        
        # Drop rows with NaN values
        df = df.dropna()
        
        # Save the modified DataFrame to a new Excel file in the destination directory
        df.to_excel(destination_file_path, index=False)
        
        print(f"Processed '{file}' and saved to '{destination_file_path}'")
    else:
        print(f"Ignored '{file}' as it is not an Excel file")
```
