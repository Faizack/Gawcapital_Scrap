```python
import os

# Specify the directory path
directory = r'c:\Users\faiza\Music\greenfi\gawcapital\src\planet-tracker\data'
directory2 = r'c:\Users\faiza\Music\greenfi\gawcapital\src\planet-tracker\DB'

# List all files in the directory
files = os.listdir(directory)

print(files)


  
```

```python
# # Iterate through each file
for filename in files:
    # Check if the filename contains '_report'
    if '_Report' in filename:
        # Create the new filename by replacing '_report' with ''
        new_filename = filename.replace('_Report', '')
        # Construct the full file paths
        old_path = os.path.join(directory, filename)
        new_path = os.path.join(directory2, new_filename)
        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed {filename} to {new_filename}")
```
